# DATA VISUALIZATION
Data science umumnya berhadapan dengan data yang sangat banyak sehingga untuk mendapatkan insight atau informasi dari kumpulan data tersebut perlu dilakukan visualisasi data. Dalam suatu perusahaan tidak semua paham untuk membaca data yang masih mentah. Untuk mempermudah pemahaman pihak lain mengenai hasil analisis maka ubahlah data ke dalam bentuk yang lebih menarik dan mudah dipahami. Beberapa jenis visualisasi data yang digunakan dalam pengolahan data adalah diagram batang, garis, lingkaran, boxplot, scatter plot, pair plot, dan heatmap correlation.

```python
#BAR CHART
MataKuliah = ['Dasar Pemrograman','Basis Data',
              'Machine Learning','Pemrograman Objek',
              'Pemrograman Web']
Jumlah_mahasiswa = [125,90,120,99,77]

plt.figure(figsize=(12,7))
plt.barh(MataKuliah, Jumlah_mahasiswa, color='blue')
plt.title('Jumlah mahasiswa pada mata kuliah di semester 3',
          size=14)
plt.xlabel('Jumlah mahasiswa', size=12)
plt.xticks(size=12)
plt.yticks(size=12)
plt.show()
```

Line chart biasanya digunakan untuk menampilkan data berbentuk time series. Artinya ada perubahan data dari waktu ke waktu.

```python
#LINE CHART
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
Nilai = [80,90,65,79,82]
ax.plot(Nilai)
plt.show()
```

Pie plot digunakan untuk menyajikan data dalam bentuk persentase, yang mana setiap potongan pie berisikan data dengan ukuran tertentu. Secara visual pie plot menarik untuk penyajian data serta memudahkan pembaca dalam memahami data tersebut.

```python
#PIE CHART
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
stock = ['Cheese','Bread','Milk','Ice Cream','Noodles','Egg']
sold = [10,50,45,15,37,113]
ax.pie(sold, labels=stock, autopct='%1.2f%%')
plt.show()
```

Hasil pengamatan:
- produk yang banyak terjual adalah 'egg'
- produk yang paling sedikit terjual adalah 'cheese'

```python
#BOXPLOT
akomodasi = (2000,1160,6100,5800,6400,900)
kamar = (1503,10732,3653,5206,2187,4781,523)
kasur = (9257,16418,6015,7832,2885,6800,786)
fig, ax = plt.subplots()
ax.boxplot((akomodasi, kamar, kasur), 
           vert=False, 
           showmeans=True,
           meanline=True, 
           labels=('Akomodasi','Kamar','Kasur'), 
           patch_artist=True, 
           medianprops={'linewidth': 2, 'color':'blue'}, 
           meanprops={'linewidth': 2, 'color':'red'})
plt.show()
```

Keterangan:
- `vert` mengatur orientasi plot menjadi horizontal saat False. Orientasi default adalah vertikal.
- `showmeans` menunjukkan mean ketikaTrue.
- meanline merepresentasikan mean pada baris ketikaTrue.
- `labels`: label data Anda.
- `patch_artist` menentukan cara menggambar grafik.
- `medianprops` menunjukkan sifat-sifat garis yang mewakili median.
- `meanprops` menunjukkan sifat-sifat garis yang mewakili mean.

```python
#MENGATASI DATA KOSONG (MISSING VALUE)
df.isna().sum()
```

```python
#mengisi nilai missing value dengan ketentuan:
#modus untuk data kategorik dan mean untuk data numerik
from statistics import mode

jenis_kelamin_mode = mode(df['jenis_kelamin'])
pekerjaan_mode = mode(df['pekerjaan']) 
is_keturunan_mode = mode(df['is_keturunan']) 
sampo_mode = mode(df['sampo'])
berat_mean = df['berat'].mean() 
gaji_mean = df['gaji'].mean()

kolom_missing = ['jenis_kelamin','pekerjaan',
                 'is_keturunan','berat','sampo','gaji']
inputan = [jenis_kelamin_mode, pekerjaan_mode,
           is_keturunan_mode, berat_mean, sampo_mode, gaji_mean]

for kolom, inputan in zip(kolom_missing, inputan): 
    df[kolom].fillna(inputan, inplace=True) 
    
#contohin cara ngisi satu kolom aja
df['jenis_kelamin'].fillna(jenis_kelamin_mode, inplace=True) 
```
